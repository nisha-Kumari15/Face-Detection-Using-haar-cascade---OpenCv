# FACE DETECTION 

```python
import cv2
alg = "haarcascade_frontalface_default.xml"
haar_cascade = cv2.CascadeClassifier(alg)

cam = cv2.VideoCapture(0)

count = 1
while count<500:
    print(count)
    _,img = cam.read()  #Read from Camera
    grayImg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)    # Color to gray
    face = haar_cascade.detectMultiScale(grayImg, 1.3, 4)   # Getting the face Coordinates
    for (x,y,w,h) in face:
        cv2.rectangle(img,(x,y),(x+w,y+h), (0,244,0),2)
        count+=1;
    cv2.imshow("Face Detection",img)
    key = cv2.waitKey(1) & 0xFF
    if key == ord("k"):
        break
cam.release()
cv2.destroyAllWindows()
```
